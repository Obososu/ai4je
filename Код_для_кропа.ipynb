```
for i in os.listdir('/content/content/labels'):
  path = '/content/content/labels/' + i
  myfile=open(path,'r')
  lines=myfile.readlines()
  pattern= "3"
  Smax = 0
  k = 0
  for line in lines:
    im = Image.open("/content/content/photo/" + i[:-4] + '.jpg')
    (width, height) = im.size
    if line[0] == pattern:
      k+=1
      c, x, y, w, h = line.split()
      x_min1 = int((float(x) - float(w)/2)*width)
      x_max1 = int(x_min1+ (float(w))*width)
      y_min1 = int((float(y) - float(h)/2)*height)
      y_max1 = int(y_min1+ (float(h))*height)
      S = (y_max1-y_min1)*(x_max1-x_min1)
      if S > Smax:
        Smax = S
        x_min = x_min1
        x_max = x_max1
        y_min = y_min1
        y_max = y_max1
  if k > 0: 
    img2 = im.crop((x_min, y_min, x_max, y_max))
    img2.save('/content/content/crop_jewel/' + i[:-4] + '_crop.jpg')
```
